# Transformer

# BERT

* BERT - языковая модель. Это идея по предобучению модели, а не архитектуры
* Обучалась 2-м задачкам - MLM и NSP
* MLM - Masked Language Modeling, восстановление маскированных слов
* NSP - Next Sentence Prediction, является ли предложение продолжением предыдущего
* В чем сила, брат? В big date-е

    * DistilBERT by HuggingFace - a supposedly smaller, faster, cheaper version of BERT that is trained from BERT, and then certain architectural aspects are removed for the sake of efficiency
    * RuBERT - обученный на русском языке

# BERT fine-tuning (Transfer learning)

* Transfer learning (перенос обучения) - берем готовую модель, по необходимости меняем архитектуру (обычно - выходной слой) и обучаем на новых данных или новой задаче
* Fine-tuning - дообучаем уже обученную нейросеть на новых данных. Может использовать как синоним transfer learning-a
* Для файнтюнинга требуется меньше данных, соответственно - меньше вычислительных ресурсов и меньше времени для обучения

# AGI

* **Artificial general intelligence** - это способность интеллектуального агента понимать или изучать любую интеллектуальную задачу, которую может выполнить человек. Синоним для "общего ИИ" и "сильного ИИ"
* Противопоставляется "слабому (узкому) ИИ" - ИИ, не предназначенному для общих когнитивных способностей. По сути, слабый ИИ это любая программа, которая предназначена для решения ровно одной задачи
* Единого определения интеллекта и интеллектуальных задач - нет, но большинство исследователей согласны с тем, что интеллект необходим для:
    * суждения - использования стратегии, решения головоломок и принятия решений в условиях неопределенности
    * представления знаний
    * планирования
    * обучения
    * общения на естественном языке
    * ощущения (видеть, слышать и т. д.)
    * результат как способность действовать (например, перемещать объекты и манипулировать ими, менять свое местоположение для исследования и т. д.)
    * объединения перечисленных навыков
* Соответственно, если система выполняет все эти действия, то она AGI
* Тест Тьюринга - если человек_1 при взаимодействии с системой и человеком_2 не может отличить систему от человека, то система - AGI

* Китайская комната - эксперимент для опровержении утверждения, что цифровая машина, наделённая «искусственным интеллектом» путём её программирования определённым образом, способна обладать сознанием в том же смысле, в котором им обладает человек. Иными словами, целью является опровержение гипотезы так называемого «сильного» искусственного интеллекта и критика теста Тьюринга
* Суть эксперимента - в закрытой комнате находится человек, не знающий китайского языка и книга, содержащая только правила преобразования одних последовательностей иероглифов в другие (без объяснения их смысла). Наблюдатель дает человеку последовательность иероглифов с вопросом, человек по книге выполняет некоторых алгоритм и выдает ответ. Предположим, что ответ всегда имеет смысл. Можем ли мы при этом считать, что человек понимает китайский?

    * Использование структурированных видов хранения информации

# Графы знаний

* Граф знаний - семантическая сеть, в которой хранится информация о разных сущностях и взаимосвязях между ними. Сущностью или «узлом»‎ графа может быть что угодно: человек, предмет, дата, концепция и любой другой материальный объект или абстрактное понятие. Предикаты или «ребра» отражают взаимосвязи и отношения между разными сущностями в графе

### А при чем тут нейронные сети?

* Генерация эмбеддингов

<a href=""><img src="https://www.researchgate.net/publication/338026955/figure/fig2/AS:837679767883777@1576729828909/llustrations-of-Translation-based-embeddings-a-TransE-and-b-TransD-embedding-method.png" alt="1" border="0"></a>

* Оптимальное использование графа

    Граф знаний используется как внешняя память, маленькая нейросеть обучена использовать этот граф для решения задачи

* Графовые нейронные сети
<a href=""><img src="https://www.researchgate.net/publication/353136116/figure/fig2/AS:1043825124458502@1625878711717/Network-architecture-of-the-GNN-model-A-series-of-message-passing-layers-MPLs-are.png" alt="1" border="0"></a>
    * Graph Convolutional Networks (GCNs) are similar to traditional CNNs. It learns features by inspecting neighboring nodes. GNNs aggregate node vectors, pass the result to the dense layer, and apply non-linearity using the activation function. In short, it consists of Graph convolution, linear layer, and non-learner activation function. There are two major types of GCNs: Spatial Convolutional Networks and Spectral Convolutional Networks.
    * Graph Auto-Encoder Networks learn graph representation using an encoder and attempt to reconstruct input graphs using a decoder. The encoder and decoders are joined by a bottleneck layer. They are commonly used in link prediction as Auto-Encoders are good at dealing with class balance.
    * Recurrent Graph Neural Networks(RGNNs) learn the best diffusion pattern, and they can handle multi-relational graphs where a single node has multiple relations. This type of graph neural network uses regularizers to boost smoothness and eliminate over-parameterization. RGNNs use less computation power to produce better results. They are used in generating text, machine translation, speech recognition, generating image descriptions, video tagging, and text summarization.
    * Gated Graph Neural Networks (GGNNs) are better than the RGNNs in performing tasks with long-term dependencies. Gated Graph Neural Networks improve Recurrent Graph Neural Networks by adding a node, edge, and time gates on long-term dependencies. Similar to Gated Recurrent Units (GRUs), the gates are used to remember and forget information in different states.

* Задачи для графовых нейросетей:
    * Graph Classification: we use this to classify graphs into various categories. Its applications are social network analysis and text classification.
    * Node Classification: this task uses neighboring node labels to predict missing node labels in a graph.
    * Link Prediction: predicts the link between a pair of nodes in a graph with an incomplete adjacency matrix. It is commonly used for social networks.
    * Community Detection: divides nodes into various clusters based on edge structure. It learns from edge weights, and distance and graph objects similarly.
    * Graph Embedding: maps graphs into vectors, preserving the relevant information on nodes, edges, and structure.
    * Graph Generation: learns from sample graph distribution to generate a new but similar graph structure

    * Биологическая - человек обучается множеству задач и может применять опыт в одних для решения других (распознавание объектов для распознавания лиц, например)

    * Научная - обучение нескольким задачам позволяет ввести в модель смещение для повышения обобщающей способности. Мы ожидаем, что для принятия решения по гипотезе может пригодиться опыт в нескольких задачах

* Parameters sharing
    * Hard - общие веса для всех задач
    * Soft - раздельные веса для всех задач + переодическое обновление весов каждой модели на основе весов других моделей
